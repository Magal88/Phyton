**We use the database of patients who have the hepatitis B virus (https://www.kaggle.com/datasets/drlucaserring/hepatitis-b-virus-levels-of-patients)  and we will use some functions from the pandas library.**

```python
import pandas as pd
```

```python
pacientes_hbv = pd.read_excel("hbv_data.xlsx")
```

```python
type(pacientes_hbv)
```

```python
print(pacientes_hbv.shape)
```

In this case we observe that the database has 100 rows and 9 columns.

If we use the head and tail commands, we can print the first and last rows of the database.

```python
print(pacientes_hbv.head(3))
```

```python
print(pacientes_hbv.tail(10))
```

```python
pacientes_hbv['age'].describe()
```

Using the describe function, we obtain statistical measurements (mean, standard deviation, minimum, maximum and quartiles). 
You can also obtain each measurement individually:

```python
pacientes_hbv['age'].min()
```

```python
pacientes_hbv['age'].max()
```

```python
pacientes_hbv['age'].mean()
```

```python
pacientes_hbv['age'].std()
```

In the following case, those patients of the female gender are filtered.

```python
pacientes_hbv[pacientes_hbv.gender == "female"]
```

In the following case, those female patients who have kidney damage are printed on the screen:

```python
pacientes_hbv[(pacientes_hbv.otherdisease== "kidney failure") & (pacientes_hbv.gender == "female")]
```

```python
Some columns can be renamed:
```

```python
pacientes_hbv = pacientes_hbv.rename(columns={'id': 'ID', 'gender': 'Gender', 'age': 'Age'})
 
print(pacientes_hbv)
```

You can also reorder columns alphabetically:

```python
pacientes_hbv =pacientes_hbv.sort_index(axis=1)
 
print(pacientes_hbv)
```
